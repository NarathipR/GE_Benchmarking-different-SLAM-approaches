```python
import pandas as pd
import matplotlib.pyplot as plt
from pyproj import Proj, transform,Transformer

transformer = Transformer.from_crs("EPSG:4326", "EPSG:26917", always_xy=True)
log_file = '100831_155323_Ins0.log'  
column_names = ['timestamp', 'some_value', 'latitude', 'longitude', 'altitude', 'value1', 'value2', 'value3', 'value4', 'value5', 'value6', 'value7']

gps_data = pd.read_csv(log_file, sep=' ', header=None, names=column_names)

latitude = gps_data['latitude']
longitude = gps_data['longitude']
altitude = gps_data['altitude']

xx, yy = transformer.transform(longitude, latitude)

plt.figure(figsize=(10, 6))
plt.scatter(xx, yy, marker='o', color='b', label='Points')
plt.title('X vs Y Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xx, yy, altitude, marker='o', color='b', label='Points')

ax.set_title('3D Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
ax.legend()
plt.show()
```
